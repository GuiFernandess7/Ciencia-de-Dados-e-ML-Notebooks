```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 100 numeros espa√ßados igualmente de 0 a 10
y = np.sin(x) # seno dos valores de x
z = np.cos(x) # cosseanos dos valores de x
```

```python
# Plot dos senos em rela√ß√£o aos valores de x
plt.plot(x, y)
plt.show()
```

```python
# Plot dos senos em rela√ß√£o aos valores de x
plt.plot(x, z)
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z)
plt.xlabel('Angle')
plt.ylabel('Value')
plt.title("Sine x Cos Waves")
plt.show()
```

```python
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

### Plot de Barras (Bar Plot)

```python
fig = plt.figure()
# Os primeiros zeros s√£o as coordenadas e os outros uns s√£o a area da figura
ax = fig.add_axes([0, 0, 1, 1]) 

# Features
languages = ["English", "French", "Spanish", "Latin", "German"]
people = [100, 50, 1500, 400, 70]

ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

### Chart de Torta (Pie Chart)

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])

# Features
languages = ["English", "French", "Spanish", "Latin", "German"]
people = [100, 50, 1500, 400, 70]

ax.pie(people, labels=languages, autopct='%1.1f') #autopct √© o ponto decimal
plt.show()
```

### Plot de pontos (Scatter Plot)

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='g')
ax.scatter(x, z, color='b')
plt.show()
```

### 3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
